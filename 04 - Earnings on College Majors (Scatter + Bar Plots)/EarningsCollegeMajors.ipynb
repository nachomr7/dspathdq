# Visualizing Earnings Based On College Majors

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

## Opening and cleaning the data

We'll start by importing the libraries needed for this project along with opening our dataset and cleaning it up a bit.

```python
import pandas as pd
import matplotlib as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
rows_dropped = raw_data_count - cleaned_data_count
print(rows_dropped,'row(s) with missing values were dropped')
```

## Visualizing and analyzing the data - Part 1

Now we'll use some scatter plots to compare the correlations between some variables and their dependency with each other.

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

We can observe how majors that are not popular have the lowest incomes, while we see a small trend that when majors are popular they make more money.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

Here there isn't a clear correlation, instead we can see how the vast majority of the records are between 5 and 12.5% of unemployment rate.

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

Similar to the `sample_size` vs `median` scatter plot, there's a weak trend that full time jobs make more money.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

In this scatter plot there's no correlation at all, so there isn't any underlying relation between these two variables.

```python
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

From these two last plots, we can see how there are more male graduates and how their incomes are higher than those of women.

## Visualizing and analyzing the data - Part 4

Finally we'll use bar plots to compare different subsets of variables and see what can we conclude from there.

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

In the first bar plot we see how engineering majors have very low share of women, while psychology majors have a very high share of women.

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

We can see how Nuclear Engineering and Clinical Psychology are two of the majors with highest unemploymen rate, while Petroleum Engineering and Metallurgical Engineering have very low ones.
